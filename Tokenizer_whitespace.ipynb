```python
# dowload the text fro Alice in Wonderland using NLTK
from nltk.corpus import gutenberg
alice = gutenberg.raw('carroll-alice.txt')
```

```python
words_alice = alice.split()
```

```python
# Split the train and test sets, using the last 1000 words as the test set.
train = words_alice[:-1000]
test = words_alice[-1000:]
```

```python
# Get the training data
x = ""
y = ""
for word in train:
  for i in range(len(word)):
    if i != len(word) - 1:
      x += word[i]
      y += "0"
    else:
      x += word[i]
      y += "1"

x
```

```python
y
```

Perform Feature Engineering:

```python
# Get all the unique characters in the training set
x_set = set(x)
sent = sorted(list(x_set)) + [" "]
```

```python
# Training data including the start and the stop symbol for the 11 gram
x_sub = x
x_sub = [" ", " ", " ", " ", " "] + list(x) + [" ", " ", " ", " ", " "]
```

```python
# Get the X matrix for training
import numpy as np

aaa = len(sent)
aa = {x:i for i,x in enumerate(sent)}

matrix_X = []

X = []
Y = []
for i in range(5, len(x_sub)-5):
  sentences = x_sub[i-5:i+6]
  a = {}
  for xx in sentences:
    a[xx] = aa[xx]
  a_len = len(sentences)

  index = []
  for s in sentences:
      index.append(a[s])
  index = np.asarray(index)
  new = np.zeros((a_len,aaa))
  new[np.arange(a_len), index] = 1
  new = np.array(new)
  matrix_X.append(new.T.flatten())

matrix_X = np.array(matrix_X)
```

```python
Y = []
for i in range(0, len(y)):
  Y.append(int(y[i]))
matrix_Y = np.array(Y)
```

```python
matrix_X[0].shape
```

```python
# Get the test data
x_test = ""
y_test = ""
for word in test:
  for i in range(len(word)):
    if i != len(word) - 1:
      x_test += word[i]
      y_test += "0"
    else:
      x_test += word[i]
      y_test += "1"
```

```python
x_test
```

```python
Y_test = []
for val in y_test:
  Y_test.append(int(val))

Y_test = np.array(Y_test)
Y_test
```

```python
x_t = [" ", " ", " ", " ", " "] + list(x_test) + [" ", " ", " ", " ", " "]
```

```python
def matrix(sent_dict, x_sent):
  aaa = len(sent_dict)
  aa = {x:i for i,x in enumerate(sent_dict)}

  matrix_X = []
  x_sub = x_sent
  X = []
  Y = []
  for i in range(5, len(x_sub)-5):
    sentences = x_sub[i-5:i+6]
    a = {}
    for xx in sentences:
      a[xx] = aa[xx]
    a_len = len(sentences)

    index = []
    for s in sentences:
        index.append(a[s])
    index = np.asarray(index)
    new = np.zeros((a_len,aaa))
    new[np.arange(a_len), index] = 1
    new = np.array(new)
    matrix_X.append(new.T.flatten())

  matrix_X = np.array(matrix_X)
  return matrix_X
```

```python
# Get the X matrix for testing
matrix_X_t = matrix(sent, x_t)
```

```python
matrix_X.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 10000)
```

```python
lr.fit(matrix_X, matrix_Y)
```

```python
y_pred = lr.predict(matrix_X_t)
```

```python
def accuracy(y_actual, y_predicted):
  N = y_actual.shape
  acc = ((y_actual == y_predicted).sum())/N
  return acc[0]
```

```python
print("The per character segmentation accuracy is:", accuracy(Y_test, y_pred))
```

```python
pred_str = ""

for i in range(len(y_pred)):
  if y_pred[i] == 0:
    pred_str += x_test[i]
  elif y_pred[i] == 1:
    pred_str += (x_test[i] + " ")
  else:
    pred_str += ""

str(pred_str)
```

```python
x_test
```

```python
pred_string = pred_str.split()
```

```python
# Index length pairs
idx = 0
idx_pairs_test = []
for val in pred_string:
  idx_pairs_test.append((idx, len(val)))
  idx += len(val)
```

```python
# Actual test string
ix = 0
ix_pairs_actual = []
for v in test:
  ix_pairs_actual.append((ix, len(v)))
  ix += len(v)
```

```python
set1 = set(ix_pairs_actual)
set2 = set(idx_pairs_test)
newList = list(set1.intersection(set2))
```

```python
tp = len(newList)
fn = len(test) - len(newList)
fp = len(pred_string) - len(newList)
```

```python
def F1_score(true_p, false_p, false_n):
  precision = true_p/(true_p + false_p)
  recall = true_p/(true_p + false_n)
  f1 = true_p/(true_p + 0.5*(false_p + false_n))
  
  return round(f1,3),round(precision,3), round(recall,3)
```

```python
print("The number of true positives:", tp)
print("The number of false positives:", fp)
print("The number of false negatives:", fn)
print("The number of true negatives:", 0)
```

```python
print("The precision on the test set:", F1_score(tp, fp, fn)[1])
print("The recall on the test set:", F1_score(tp, fp, fn)[2])
print("The F1 Score:", F1_score(tp, fp, fn)[0])
```
